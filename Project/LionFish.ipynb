```python
%pylab inline

import os
import cv2
import numpy as np
import re
import pylab as pl
from fnmatch import fnmatch
from IPython.core.display import Image
```

```python

faceCascade = cv2.CascadeClassifier("newcascade23.xml")  #Let's load the required XML classifiers.

```

```python
img = cv2.imread("lion1.JPG")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)  #load input image in grayscale mode
```

```python
faces = faceCascade.detectMultiScale(gray, 1.0009, 5)   #to find faces and it is defined like this
print(faces)                                                   #If faces are found, it returns the positions of detected faces as Rect(x,y,w,h) which means x,y is the starting point

for (x,y,w,h) in faces:
    img = cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
    
imshow(img)
#finding on grayscale image but plotting on color image
```
